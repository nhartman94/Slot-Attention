# ISA loss curves : vary $\alpha$

**Goal:** Get the loss curves for a few $\alpha$ getting varied.

```python
import matplotlib.pyplot as plt
import json
import pandas as pd
import os
```

```python
os.listdir('../../code/models/isa-alpha2/')
```

```python
with open('../../code/models/isa-alpha1/loss.json') as f:
    df1 = pd.DataFrame(json.load(f))

with open('../../code/models/isa-alpha2/loss.json') as f:
    df2 = pd.DataFrame(json.load(f))
```

```python
df1.plot()
plt.xlabel('Iter',fontsize=20)
plt.ylabel('Loss',fontsize=20)
plt.title('alpha = 1')
```

```python
df2.plot()
plt.xlabel('Iter',fontsize=20)
plt.ylabel('Loss',fontsize=20)
plt.title('alpha = 2')
```

```python
for df,c,a in zip([df1,df2],['darkorange','hotpink'],[1,2]):
    plt.plot(df['bce'].rolling(100).mean(),color=c,label=r'$\alpha$ = '+f'{a}')
    
plt.xlabel('Iter',fontsize=20)
plt.ylabel('BCE Loss',fontsize=20)
plt.legend(fontsize=14)
plt.title(r'L_tot = L_bce + $\alpha$ L_mse',loc='right')
# plt.xlim(0,12000)
plt.show()
```
